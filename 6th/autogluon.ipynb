```python
df=pd.read_csv('./csv/continuous_prep.csv', index_col =0)
df.head()
```

```python
from sklearn.model_selection import train_test_split
from autogluon.tabular import TabularPredictor
 

train, test =  train_test_split(df, random_state =77,stratify=df['default_tf'])
 
# autogluon 
predictor = TabularPredictor(label='default_tf').fit(train_data=train)
```

```python
results = predictor.fit_summary()
```

```python
target = test['default_tf']
test1 = test.drop('default_tf', axis=1)
```

```python
prediction=predictor.predict(test1)
```

```python
import sklearn.metrics as metrics
cm = metrics.confusion_matrix(target, prediction)
print(cm)
```

```python
predictor.get_model_best()
```

```python
predictor1 = TabularPredictor(label='default_tf', eval_metric='precision').fit(train)
```

```python
predictor1.leaderboard(test, silent=True)
```

```python
prediction1=predictor1.predict(test1)
```

```python
cm1 = metrics.confusion_matrix(target, prediction1)
print(cm1)
```

```python
predictor2 = TabularPredictor(label='default_tf', eval_metric='recall').fit(train)
```

```python
predictor2.leaderboard(test, silent=True)
```

```python
prediction2=predictor2.predict(test1)
```

```python
cm2 = metrics.confusion_matrix(target, prediction2)
print(cm2)
```
